##### import lib and read csv

```python
import pandas as pd 

worldpop = pd.read_csv('/workspaces/bts_jewel/data/world_population.csv', index_col=False)
btstrends = pd.read_csv('/workspaces/bts_jewel/data/bts_trends.csv', index_col=False)
```

##### performed left join to merge the 2 csv 

```python
df=btstrends.merge(worldpop, on='Country', how='left')
```

#### drop the unwanted columns 

```python
df2=df.drop(df.columns[[0,4]], axis=1,)
df2
```

#### sort the data by Population (2020)

```python
df3=df2.sort_values(by=['Population (2020)'], ascending=False)
df3
```

```python
df3.rename(columns={"Med. Age":"Med_Age"}, inplace=True)
df3
```

#### filter the data by Median Age

```python
df4=df3.query('18 <= Med_Age <= 25')
df4
```

#### sort the data based on bts interest per region

```python
df5=df4.sort_values(by=['bts'], ascending=False)
df5.nlargest(6, 'bts')
```
